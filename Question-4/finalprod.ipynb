```python

dir = '/content/dataset'
cats = ['apples', 'grapes', 'papaya', 'watermelon']
# imported necessary library
import cv2
import os
import random
import matplotlib.pyplot as plt
data = []
# google colab images path
img_path =  '/content/apples/apples/apples fruit'
import glob
# iterated all the images
images = glob.glob(img_path + '/*.jpg')

for file in images:
   img_arr= cv2.imread(file)
   img_arr = cv2.resize(img_arr, (100,100))
   data.append([img_arr, 0])
   
# preprocessed the image to 100 size
img_path =  '/content/images/grapes/grapes fruit'
import glob
images = glob.glob(img_path + '/*.jpg')
for file in images:
   img_arr= cv2.imread(file)
   img_arr = cv2.resize(img_arr, (100,100))
   data.append([img_arr, 1])

img_path =  '/content/images/papaya/papaya fruit'
import glob
images = glob.glob(img_path + '/*.jpg')
for file in images:
   img_arr= cv2.imread(file)
   # preprocessed the image to 100 size
   img_arr = cv2.resize(img_arr, (100,100))
   data.append([img_arr, 2])

img_path =  '/content/watermelon/papaya/watermelon fruit'
import glob
images = glob.glob(img_path + '/*.jpg')
for file in images:
   img_arr= cv2.imread(file)
   # preprocessed the image to 100 size
   img_arr = cv2.resize(img_arr, (100,100))
   data.append([img_arr, 3])
```

```python
print(len(data))
```

```python
# shuffled the data
random.shuffle(data)
```

```python
# printed the sample data
data[1]
```

```python
X = []
Y = []

for features, labels in data:
  X.append(features)
  Y.append(labels)
```

```python
import numpy as np
X = np.array(X)
Y = np.array(Y)
```

```python
print(len(X))
print(len(Y))
```

```python
from sklearn.model_selection import train_test_split
# splitted the data into train and test daata
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.33, random_state=1)
```

```python
print(len(X_train))
print(len(y_train))
```

```python
print(X_train[0])
```

```python
X_train = X_train.astype('float32')
X_test = X_test.astype('float32')
X_train = X_train / 255.0
X_test = X_test / 255.0
```

```python
# printed the sample ttrian data
print(X_train[0])
```

```python
import numpy
from keras.datasets import cifar10
from keras.models import Sequential
from keras.layers import Dense, Flatten, Dropout, Input, BatchNormalization
from keras.constraints import maxnorm
from keras.models import Model
from keras.optimizers import *
from keras.layers.convolutional import Conv2D, MaxPooling2D
from keras.utils import np_utils
from keras import backend as K
# categorical encoding
y_train = np_utils.to_categorical(y_train)
y_test = np_utils.to_categorical(y_test)
num_classes = y_test.shape[1]
```

```python
model1 = Sequential()
# created the sequential model with filter of size 3*3 and 100 filter
model1.add(Conv2D(100, (3, 3), input_shape=(100,100, 3), padding='same', activation='relu'))
model1.add(Dropout(0.2))
# created the sequential model with filter of size 3*3 and 100 filter
model1.add(Conv2D(100, (3, 3), activation='relu', padding='same'))
model1.add(MaxPooling2D(pool_size=(2, 2)))
# created the sequential model with filter of size 3*3 and 64 filter
model1.add(Conv2D(64, (3, 3), activation='relu', padding='same'))
model1.add(Dropout(0.2))
# created the sequential model with filter of size 3*3 and 64 filter
model1.add(Conv2D(64, (3, 3), activation='relu', padding='same'))
model1.add(MaxPooling2D(pool_size=(2, 2)))
# created the sequential model with filter of size 3*3 and 128 filter
model1.add(Conv2D(128, (3, 3), activation='relu', padding='same'))
model1.add(Dropout(0.2))
# created the sequential model with filter of size 3*3 and 128 filter
model1.add(Conv2D(128, (3, 3), activation='relu', padding='same'))
model1.add(MaxPooling2D(pool_size=(2, 2)))

model1.add(Flatten())
model1.add(Dropout(0.2))
# model layer with 1024 neurons
model1.add(Dense(1024, activation='relu'))
model1.add(Dropout(0.2))
# model layer with 512 neurons
model1.add(Dense(512, activation='relu'))
model1.add(Dropout(0.2))
model1.add(Dense(num_classes, activation='softmax'))
```

```python
epochs = 10
lrate = 0.002
adam = tf.keras.optimizers.Adam(learning_rate=lrate)
model1.compile(loss='categorical_crossentropy', optimizer=adam, metrics=['accuracy'])
```

```python
print(model1.summary())
```

```python
# trained the model
history = model1.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=epochs, batch_size=128, verbose=1)
```

```python
# model evaluate
scores1 = model1.evaluate(X_test, y_test, verbose=0)
print("Accuracy: %.2f%%" % (scores1[1]*100))
```
